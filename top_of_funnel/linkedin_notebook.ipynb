```python
# Read in LinkedIn campaign data
campaigns_df = combine_files_in_dir(os.path.join("Gtmhub Ads Data", "LinkedIn Ads", "campaigns"))
```

```python
# Rename main datetime field
campaigns_df.rename(columns={"Start Date (in UTC)": "start_date"}, inplace=True)

# Drop time portion from datetime
campaigns_df["start_date"] = campaigns_df["start_date"].dt.date
```

```python
# Read in hubspot data
hs_contacts = pd.read_csv("hubspot_contacts.csv")
```

```python
hs_contacts.head()
```

```python
# Take each campaign__paid in propertiesWithHistory and create a new row based on it
hs_contact_campaign = []
count = 0

for index, row in hs_contacts.iterrows():
    # print(row)

    properties_with_history = row["propertiesWithHistory"].replace("'", '"')

    history = json.loads(properties_with_history)

    for campaign_paid, campaign_source in zip(history["campaign___paid"], history["campaign_source"]):
        # print(item)
        existing_row_data = {
            header: row[header]
            for header in row.index
            if "properties" not in header
        }
        
        # Add prefixes to avoid naming collusions
        paid = {f"campaign_paid.{key}": value for key, value in campaign_paid.items()}
        source = {f"campaign_source.{key}": value for key, value in campaign_source.items()}

        hs_contact_campaign.append({**existing_row_data, **paid, ** source})

hs_contact_campaign_df = pd.DataFrame(hs_contact_campaign)
```

```python
hs_contact_campaign_df.columns.tolist()
```

```python
# Reference the two columns containing Salesforce IDs
contact_ids = hs_contact_campaign_df["salesforcecontactid"]
lead_ids = hs_contact_campaign_df["salesforceleadid"]

# Create a new column to handle the Salesforce join
hs_contact_campaign_df["salesforce_id"] = contact_ids.combine_first(lead_ids)

# Drop the old id columns
hs_contact_campaign_df.drop(columns=["salesforcecontactid", "salesforceleadid"], inplace=True)
```

```python
# hs_contact_campaign_df["createdAt"] = pd.to_datetime(hs_contact_campaign_df["createdAt"])
hs_contact_campaign_df["campaign.created_at"] = hs_contact_campaign_df["campaign_paid.timestamp"].map(datetime_parser.parse)
# hs_contact_campaign_df["createdAt"] = hs_contact_campaign_df["createdAt"].tz_localize(None)
```

```python
# Filter down data to 2020 and later
hs_contact_campaign_df = hs_contact_campaign_df[
    hs_contact_campaign_df["campaign.created_at"] >= datetime(year=2020, month=1, day=1).isoformat()
]
```

```python
# Remove rows that don't have a value for campaign___paid
hs_contact_campaign_df = hs_contact_campaign_df[
    ~hs_contact_campaign_df["campaign_paid.value"].isnull()
]
```

```python
hs_contact_campaign_df = hs_contact_campaign_df[
    hs_contact_campaign_df["campaign_source.value"].str.contains("linkedin", case=False)
]
```

```python
# Read in Salesforce data
sf_df = pd.read_csv("salesforce_prospects_data_for_ads.csv")
```

```python
# Merge Hubspot with Salesforce
hs_sf_contact_campaign_df = hs_contact_campaign_df.merge(
    sf_df,
    how="left",
    left_on="salesforce_id",
    right_on="Id"
)
```

```python
hs_sf_contact_campaign_df.head()
```

```python
hs_sf_contact_campaign_df["is_mql"] = hs_sf_contact_campaign_df["BecameMQLDate"].map(lambda x: 1 if pd.notnull(x) else 0)
```

```python
# Create new columns
hs_sf_contact_campaign_df["has_opportunity"] = hs_sf_contact_campaign_df["new_business_count"].map(lambda x: 1 if pd.notnull(x) and x > 0 else 0)
hs_sf_contact_campaign_df["has_meeting"] = hs_sf_contact_campaign_df["event_count"].map(lambda x: 1 if pd.notnull(x) and x > 0 else 0)
```

```python
# Drop time portion of datetime field
hs_sf_contact_campaign_df["campaign.created_at"] = hs_sf_contact_campaign_df["campaign.created_at"].dt.date
```

```python
day_campaign = hs_sf_contact_campaign_df.groupby(["campaign.created_at", "campaign_paid.value"]).agg(
    {"id": "count", "new_business_count": "sum", "event_count": "sum", "is_mql": "sum"}).reset_index()
day_campaign.head()
```
